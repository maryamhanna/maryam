```python
import csv
import os
import numpy as np
import re
mypath = "/home/ket/maryam/imgaug/All/"
dirs = os.listdir(mypath)
num_images = len(dirs)
names = []
cars = []
distance = []
rl_view = []
angle = []
dataug = []
print (num_images)
for filename in os.listdir(mypath):
    #print(filename)
    if filename is not None:
        if filename.endswith('.JPG'):
            file_name = mypath + filename 
            names.append(file_name)
            filename = re.split('[_ .]', filename)
            #print(filename)
            
            if 'car1' in filename:
                cars.append('car1')
            elif 'car2' in filename:
                cars.append('car2')
            elif 'car3' in filename:
                cars.append('car3')
            elif 'car4' in filename:
                cars.append('car4')
            elif 'car5' in filename:
                cars.append('car5')
            elif 'car6' in filename:
                cars.append('car6')

            if '2' in filename: 
                distance.append('2')
            elif '4' in filename: 
                distance.append('4')
            elif '6' in filename: 
                distance.append('6')
            elif '8' in filename: 
                distance.append('8')
            elif '10' in filename: 
                distance.append('10')
            elif '15' in filename: 
                distance.append('15')
            elif '20' in filename: 
                distance.append('20')
            elif '25' in filename: 
                distance.append('25')
            elif '30' in filename: 
                distance.append('30')

            if '0' in filename:
                rl_view.append('none')
            elif ((('30L' or '45L') and ('flipped')) or ('30R' or '30RS' or '45R')) in filename:
                rl_view.append('left')
            elif ((('30R' or '45R') and ('flipped')) or ('30L' or '45L')) in filename:
                rl_view.append('right')  
            
            if '0' in filename:
                angle.append('0')
            elif ('30L' or '30R' or '30RS') in filename:
                angle.append('30')
            elif ('45L' or '45R') in filename:
                angle.append('45')

            if ('flipped' and 'salt' and 'rotated45') in filename:
                dataug.append('sp_rot45_flip')
            elif ('flipped' and 'salt' and 'rotated30') in filename:
                dataug.append('sp_rot30_flip')
            elif ('flipped' and 'salt' and 'rotated15') in filename:
                dataug.append('sp_rot15_flip')
            elif ('salt' and 'rotated45') in filename:
                dataug.append('sp_rot45')
            elif ('salt' and 'rotated30') in filename:
                dataug.append('sp_rot30')
            elif ('salt' and 'rotated15') in filename:
                dataug.append('sp_rot15')
            elif ('flipped' and 'rotated45') in filename:
                dataug.append('rot45_flip')
            elif ('flipped' and 'rotated30') in filename:
                dataug.append('rot30_flip')
            elif ('flipped' and 'rotated15') in filename:
                dataug.append('rot15_flip')
            elif ('rotated45') in filename:
                dataug.append('rot45')
            elif ('rotated30') in filename:
                dataug.append('rot30')
            elif ('rotated15') in filename:
                dataug.append('rot15')
            elif ('flipped') in filename:
                dataug.append('flip')
            else:
                dataug.append('none')           
print('Done setting columns')

data_input = []

num = len(names)
print(num)
num = len(cars)
print(num)
num = len(distance)
print(num)
num = len(rl_view)
print(num)
num = len(angle)
print(angle)
print(num)
num = len(dataug)
print(num)
for i in range(num):
    #print(i)
    row = []
    row.append(names[i])
    row.append(cars[i])
    row.append(distance[i])
    row.append(rl_view[i])
    row.append(angle[i])
    row.append(dataug[i])
    data_input.append(row)
#print(data_input)

print ('Done setting matrix')
f = open('plates.csv', 'w')
with f:
    writer = csv.writer(f)
    writer.writerow(["Path", "Car", "Distance", "RL_View", "Angle", "Data-Aug"])
    length = len(data_input)
    for i in range(length):
        writer.writerow(data_input[i])
    print('Done saving matrix')

    
```
