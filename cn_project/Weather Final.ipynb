```python
### Project for Course: Computer Networks (06-88-556-1)
### Submitted to: Dr. Kemal Tepe
### Submitted by: Maryam Hanna & Maryam Eshaghi
### Submission date: November 27, 2018
###############################################################################
# Choose to write a program that retrieves weather infromation from the
# internet (a website) and deplay it on the screenself.
# The user will input desired country and city, and the system will provide the
# current the temperature (in Celsius), and the current weather condition
# (sunny, rainy, windy, etc.). In-addition, this code is designed in a way not
# to crash easily.
#
# Used resourses and knowledge from the following sites:
# 1) https://www.dataquest.io/blog/web-scraping-beautifulsoup/
# 2) https://www.digitalocean.com/community/tutorials/how-to-work-with-web-
#    ...data-using-requests-and-beautiful-soup-with-python-3
# 3) http://docs.python-requests.org/en/master/user/quickstart/
# 4) https://beautiful-soup-4.readthedocs.io/en/latest/
#
###############################################################################

# imported the following libraries to be able to access html skeleton of
# websites
from requests import get
from bs4 import BeautifulSoup
import requests

## Function input_cc
# input: none
# output: string named location
# Allows the user input desired country and city. The user may input the data
# in either uppercase or lowercase (or both), and add space before or after
# their input and the code will not crash.
def input_cc():
    country = input('\n\nEnter country: ')
    country = country.lstrip().lower().rstrip()
    city = input('Enter city: ')
    city = city.lstrip().lower().rstrip()
    location = str(country) + "/" + str(city)
    print(location)
    return location

## Function main_code
# input: none
# output: string url
# Calls input_cc function to attain loaction the user wants the data from. The
# setup URL is from www.timeanddate.com website; provided data like weather and
# time of cities all over the world. It covers most cities, but not all.
# In-addition, in this function, it will attain the url status code and displays
# it to the user. If the status code is anything other than 200 (ok), the code
# will display error and loop for user to try to input a new location (instead
# of crashing).
def main_code():
    location = input_cc()
    url = 'https://www.timeanddate.com/weather/' + location
    response = requests.head(url).status_code
    print('\nweb status:', response)
    while response != 200:
        print('Error, please input country and city again')
        location = input_cc()
        url = 'https://www.timeanddate.com/weather/' + location
        response = requests.head(url).status_code
        print('\nweb status:', response)
    return url

## Function get_temp
# input: string called url
# output: array of strings called data
# From the given url, it attains the website's html text. Using bs4, it finds
# the "tree-branch" that contains the weather data. In-addition, in the same
# branch, it contains the weather-conditions, which it saves. The temperature
# and weather-conditions are saved into variables as string, which is later
# saved into the data array.
def get_temp(url):
    data = []
    web_html = get(url)
    soup = BeautifulSoup(web_html.text, 'html.parser')
    temp = soup.find('div', class_= 'h2')
    weather_condition = temp.findNext('p')
    temp = temp.text
    data.append(temp)
    weather_condition = weather_condition.text
    data.append(weather_condition)
    return data

## Program's operating main_code
# By using flags, the program is in an infinte loop, unless stated otherwise
# by the user's input. The user has three options while in the middle of the
# program: refresh, restart, quit. If the user inputs refresh, it will give
# refreshed data of the temperature of the same place. If the user inputs
# restart, they will be able to input a new loaction they want to the
# temperature for. If the uer inputs quit, the program will end.
quit_flag = 0
while quit_flag == 0:
    restart_flag = 0
    url = main_code()
    while restart_flag == 0:
        data = get_temp(url)
        print ("\ncurrent temperature: ", data[0])
        print ("\nweather conditions: ", data[1])
        refresh_flag = 0
        while refresh_flag == 0:
            user_input = input('\nPlease type what next (refresh, restart, quit): ')
            if user_input == 'refresh':
                refresh_flag = 1
            elif user_input == 'restart':
                restart_flag = 1
                refresh_flag = 1
            elif user_input == 'quit' :
                quit_flag = 1
                restart_flag = 1
                refresh_flag = 1
```
